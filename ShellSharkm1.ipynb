```python
import customtkinter as ctk
from tkinter import messagebox
import random
import logging
from datetime import datetime
from passlib.hash import pbkdf2_sha256
import time


logging.basicConfig(level=logging.INFO, format='%(asctime)s - [SHARKSHELL_SENTINEL] - %(message)s')

ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

class BlueFinBank(ctk.CTkToplevel):
    """High-Fidelity Private Banking Node"""
    def __init__(self, parent):
        super().__init__(parent)
        self.title("Blue Fin Private Wealth - Secure Terminal")
        self.geometry("900x650")
        self.configure(fg_color="#0a192f")
        self.attributes("-topmost", True)
        
       
        self.sidebar = ctk.CTkFrame(self, width=200, corner_radius=0, fg_color="#0d253f")
        self.sidebar.pack(side="left", fill="y")
        ctk.CTkLabel(self.sidebar, text="BLUE FIN", font=("Arial", 22, "bold"), text_color="#3498db").pack(pady=30)
        
        for btn in ["Dashboard", "Accounts", "Transfers", "Investments"]:
            ctk.CTkButton(self.sidebar, text=btn, fg_color="transparent", anchor="w").pack(fill="x", padx=20, pady=5)

     
        self.content = ctk.CTkFrame(self, fg_color="transparent")
        self.content.pack(side="right", fill="both", expand=True, padx=30, pady=30)

      
        header = ctk.CTkFrame(self.content, fg_color="transparent")
        header.pack(fill="x", pady=(0, 20))
        ctk.CTkLabel(header, text="Welcome, Mark", font=("Arial", 24, "bold")).pack(side="left")
        ctk.CTkLabel(header, text=f"Node: MO-USA | {datetime.now().strftime('%H:%M')}", text_color="gray").pack(side="right")

        card = ctk.CTkFrame(self.content, fg_color="#1f6aa5", height=120, corner_radius=15)
        card.pack(fill="x", pady=10)
        card.pack_propagate(False)
        ctk.CTkLabel(card, text="TOTAL ASSETS", font=("Arial", 12, "bold")).pack(pady=(15, 0))
        ctk.CTkLabel(card, text="$1,245,800.50", font=("Arial", 36, "bold")).pack()

    
        ctk.CTkLabel(self.content, text="Recent Transactions", font=("Arial", 16, "bold")).pack(anchor="w", pady=(20, 10))
        ledger = ctk.CTkTextbox(self.content, height=250, fg_color="#0d253f", font=("Courier", 12))
        ledger.pack(fill="both", expand=True)
        
        log_data = [
            "2026-03-23 | Wire Transfer IN    | +$15,000.00",
            "2026-03-22 | Server Hosting Fee   | -$1,240.20",
            "2026-03-21 | SharkShell Premium   | -$199.99",
            "2026-03-21 | Dividends Received   | +$2,100.00"
        ]
        ledger.insert("0.0", "\n".join(log_data))
        ledger.configure(state="disabled")

        ctk.CTkButton(self.content, text="Secure Logout", command=self.destroy, fg_color="#c0392b").pack(pady=20)

class SharkShellApp(ctk.CTk):
    def __init__(self):
        super().__init__()
        self.title("SharkShell Enterprise Security")
        self.geometry("750x850")
        self.configure(fg_color="#050a14")

        self._user_db = {"mark123": pbkdf2_sha256.hash("1234")}
        self.session_history = []
        self.failed_attempts = 0
        self.honeypot_enabled = True

        self.main_container = ctk.CTkFrame(self, corner_radius=20, fg_color="#0a192f")
        self.main_container.pack(padx=40, pady=40, fill="both", expand=True)

        self.render_login()

    def render_login(self):
        self.configure(fg_color="#050a14")
        for w in self.main_container.winfo_children(): w.destroy()
        
        ctk.CTkLabel(self.main_container, text="🦈 🔒", font=("Arial", 60)).pack(pady=(60, 10))
        ctk.CTkLabel(self.main_container, text="SHARKSHELL", font=("Arial", 30, "bold")).pack(pady=5)
        
        self.user_input = ctk.CTkEntry(self.main_container, placeholder_text="Username", width=300, height=45)
        self.user_input.pack(pady=(40, 10))
        self.pass_input = ctk.CTkEntry(self.main_container, placeholder_text="Password", show="*", width=300, height=45)
        self.pass_input.pack(pady=10)

        self.hp_field = ctk.CTkEntry(self.main_container, width=0, height=0)
        self.hp_field.place(x=-1000, y=-1000)

        ctk.CTkButton(self.main_container, text="AUTHORIZE", command=self.handle_login, width=300, height=50).pack(pady=40)

    def handle_login(self):
        u, p = self.user_input.get(), self.pass_input.get()
        if self.honeypot_enabled and self.hp_field.get():
            self.destroy()
            return

        hashed = self._user_db.get(u)
        if hashed and pbkdf2_sha256.verify(p, hashed):
            self.session_history.insert(0, {"device": "Admin-Workstation", "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")})
            self.show_hub()
        else:
            self.failed_attempts += 1
            if self.failed_attempts >= 5: self.trigger_deception()
            else: messagebox.showerror("Denied", "Authentication Failed")

    def show_hub(self):
        for w in self.main_container.winfo_children(): w.destroy()
        
        tabs = ctk.CTkTabview(self.main_container, width=650, height=650)
        tabs.pack(padx=20, pady=20)
        tabs.add("Services")
        tabs.add("Login History")
        tabs.add("Settings")

 
        ctk.CTkLabel(tabs.tab("Services"), text="AUTHORIZED NODES", font=("Arial", 20, "bold")).pack(pady=20)
        nodes = [
            ("💰 Blue Fin Private Bank", self.bank_mfa),
            ("☁️ Private Cloud Vault", lambda: messagebox.showinfo("Cloud", "Accessing encrypted storage...")),
            ("🛋️ Furniture Biz Manager", lambda: messagebox.showinfo("Business", "Loading inventory...")),
            ("🖼️ Photo Vault", lambda: messagebox.showinfo("Media", "Unlocking gallery...")),
            ("🎮 Game Archive", lambda: messagebox.showinfo("Games", "Launching sandbox..."))
        ]
        for name, cmd in nodes:
            ctk.CTkButton(tabs.tab("Services"), text=name, width=350, height=40, command=cmd).pack(pady=8)

   
        hist_box = ctk.CTkTextbox(tabs.tab("Login History"), width=500, height=400, font=("Courier", 12))
        hist_box.pack(pady=10)
        hist_box.insert("0.0", f"{'DEVICE':<25} | {'TIMESTAMP':<20}\n" + "-"*50 + "\n")
        for entry in self.session_history:
            hist_box.insert("end", f"{entry['device']:<25} | {entry['time']:<20}\n")
        hist_box.configure(state="disabled")

      
        ctk.CTkLabel(tabs.tab("Settings"), text="ADMIN CONTROLS", font=("Arial", 18, "bold")).pack(pady=20)
        hp_switch = ctk.CTkSwitch(tabs.tab("Settings"), text="Active Honeypot Deception", 
                                   command=lambda: setattr(self, 'honeypot_enabled', not self.honeypot_enabled))
        if self.honeypot_enabled: hp_switch.select()
        hp_switch.pack(pady=10)

        ctk.CTkButton(self.main_container, text="Logout", command=self.render_login, fg_color="#c0392b").pack(pady=10)

    def bank_mfa(self):
        otp = str(random.randint(100000, 999999))
        messagebox.showinfo("MFA Required", f"Blue Fin Security Code: {otp}")
        user_otp = ctk.CTkInputDialog(text="Enter 6-digit verification code:", title="Step-Up Auth").get_input()
        if user_otp == otp: BlueFinBank(self)
        else: messagebox.showerror("Auth Failed", "Invalid MFA token.")

    def trigger_deception(self):
        for w in self.main_container.winfo_children(): w.destroy()
        ctk.CTkLabel(self.main_container, text="[!] SECURITY BYPASS DETECTED\nLOADING MASTER_VAULT_01", text_color="#2ecc71", font=("Courier", 16)).pack(pady=50)
        
        txt = ctk.CTkTextbox(self.main_container, width=550, height=250, font=("Courier", 11), fg_color="#1a1a1a")
        txt.pack(pady=20)
        fake_data = ["DB_REF | USER_ID | CLEAR_PASS | CARD_NUM", "-"*50, "00921  | admin_tp| P@ssw0rd1! | 4532...8812", "00922  | ceo_mark| Shark77!   | 5105...3301"]
        txt.insert("0.0", "\n".join(fake_data))
        txt.configure(state="disabled")
        
        self.dl_btn = ctk.CTkButton(self.main_container, text="📥 DOWNLOAD ALL RECORDS (.JSON)", fg_color="#27ae60", command=self.fake_download)
        self.dl_btn.pack(pady=20)

    def fake_download(self):
        self.dl_btn.configure(state="disabled", text="Exfiltrating...")
        p = ctk.CTkProgressBar(self.main_container, width=400)
        p.pack(pady=10)
        for i in range(1, 11):
            time.sleep(0.3)
            p.set(i/10)
            self.update()
        
        for w in self.main_container.winfo_children(): w.destroy()
        self.configure(fg_color="#4b0000")
        ctk.CTkLabel(self.main_container, text="⚠️ SYSTEM PROTOCOL: DELTA-9 ACTIVATED", font=("Arial", 20, "bold")).pack(pady=(100, 20))
        ctk.CTkLabel(self.main_container, text="UNAUTHORIZED ACCESS FLAGGED.\nYOUR IP AND DEVICE SIGNATURE LOGGED.", font=("Courier", 14)).pack()
        self.after(5000, self.destroy)

if __name__ == "__main__":
    app = SharkShellApp()
    app.mainloop()
```
