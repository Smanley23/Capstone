```python
import customtkinter as ctk
from tkinter import messagebox
import random
import logging
from datetime import datetime
from passlib.hash import pbkdf2_sha256
import time
import threading

logging.basicConfig(level=logging.INFO, format='%(asctime)s - [SHARKSHELL_SENTINEL] - %(message)s')
ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

class PhotoVaultPanel(ctk.CTkFrame):
    def __init__(self, parent, back_callback):
        super().__init__(parent, fg_color="transparent")
        self.back_callback = back_callback
        self._build()

    def _build(self):
        top_bar = ctk.CTkFrame(self, fg_color="#1a3a5c", height=60, corner_radius=10)
        top_bar.pack(fill="x", pady=(0, 10))
        top_bar.pack_propagate(False)

        ctk.CTkButton(top_bar, text="← Back", width=80, fg_color="transparent",
                      hover_color="#2c5aa0", command=self.back_callback).pack(side="left", padx=10, pady=10)
        ctk.CTkLabel(top_bar, text="🖼️  PHOTO VAULT",
                     font=("Arial", 18, "bold"), text_color="white").pack(side="left", padx=10)

        content = ctk.CTkFrame(self, fg_color="#0d1b2a", corner_radius=10)
        content.pack(fill="both", expand=True)

        sidebar = ctk.CTkFrame(content, width=240, fg_color="#0a192f", corner_radius=10)
        sidebar.pack(side="left", fill="y", padx=(10, 5), pady=10)
        sidebar.pack_propagate(False)

        ctk.CTkLabel(sidebar, text="Albums", font=("Arial", 13, "bold"), text_color="#3498db").pack(pady=15)
        
        albums = [
            ("🏢 Business Operations", "hidden"),
            ("💼 Work Projects", "hidden"),
            ("🔵 Blueprints & Designs", "hidden"),
            ("👨‍👩‍👧‍👦 Family", "visible"),
            ("🎉 Events & Celebrations", "visible"),
            ("✈️ Travel", "visible"),
            ("🌅 Vacations", "visible"),
        ]
        
        for album_name, visibility in albums:
            badge = "[HIDDEN]" if visibility == "hidden" else ""

            ctk.CTkButton(sidebar, text=f"{album_name} {badge}", fg_color="transparent", text_color="#aab7b8",
                          anchor="w", hover_color="#1a2b3c", height=38, 
                          command=lambda n=album_name: self._show_album(n)).pack(fill="x", padx=10, pady=4)

        scroll = ctk.CTkScrollableFrame(content, fg_color="transparent")
        scroll.pack(side="right", fill="both", expand=True, padx=(5, 10), pady=10)
        self.scroll_frame = scroll
        
        self._show_album("👨‍👩‍👧‍👦 Family")

    def _show_album(self, album_name):
        for widget in self.scroll_frame.winfo_children():
            widget.destroy()
        
        album_data = {
            "🏢 Business Operations": [
                ("Board Meeting 2026-Q1", "12 photos", "#c0392b"),
                ("Product Launch Event", "48 photos", "#c0392b"),
                ("Facility Inspection", "23 photos", "#c0392b"),
                ("Conference Presentation", "15 photos", "#c0392b"),
            ],
            "💼 Work Projects": [
                ("Q2 Development Cycle", "67 photos", "#c0392b"),
                ("Team Building Workshop", "34 photos", "#c0392b"),
                ("Client Site Visit", "28 photos", "#c0392b"),
                ("Office Renovation", "41 photos", "#c0392b"),
            ],
            "🔵 Blueprints & Designs": [
                ("Building Schematic A-001", "technical", "#3498db"),
                ("Floor Plan Level 2", "technical", "#3498db"),
                ("HVAC Design Layout", "technical", "#3498db"),
                ("Electrical Schematics", "technical", "#3498db"),
                ("Network Infrastructure", "technical", "#3498db"),
            ],
            "👨‍👩‍👧‍👦 Family": [
                ("Summer 2025 Memories", "89 photos", "#27ae60"),
                ("Holiday 2025", "56 photos", "#27ae60"),
                ("Weekend Adventures", "42 photos", "#27ae60"),
                ("Birthday Celebrations", "78 photos", "#27ae60"),
            ],
            "🎉 Events & Celebrations": [
                ("Wedding 2024", "234 photos", "#e74c3c"),
                ("New Year Party", "67 photos", "#e74c3c"),
                ("Corporate Gala", "143 photos", "#e74c3c"),
                ("Summer BBQ", "89 photos", "#e74c3c"),
            ],
            "✈️ Travel": [
                ("Japan Trip 2025", "234 photos", "#f39c12"),
                ("Europe Adventure", "167 photos", "#f39c12"),
                ("Caribbean Cruise", "145 photos", "#f39c12"),
            ],
            "🌅 Vacations": [
                ("Beach Week 2025", "98 photos", "#2980b9"),
                ("Mountain Retreat", "76 photos", "#2980b9"),
                ("Ski Resort 2024", "123 photos", "#2980b9"),
            ],
        }
        
        album_title = ctk.CTkLabel(self.scroll_frame, text=album_name, 
                                  font=("Arial", 18, "bold"), text_color="white")
        album_title.pack(anchor="w", padx=10, pady=(10, 20))
        
        if album_name in album_data:
            for item_name, item_info, color in album_data[album_name]:
                self._add_album_item(item_name, item_info, color)
    
    def _add_album_item(self, name, info, color):
        item = ctk.CTkFrame(self.scroll_frame, fg_color="#0a192f", height=60, corner_radius=8)
        item.pack(fill="x", pady=4)
        item.pack_propagate(False)
        
        ctk.CTkLabel(item, text=name, font=("Arial", 13, "bold"), text_color="white").pack(side="left", padx=15, pady=10)
        ctk.CTkLabel(item, text=info, font=("Arial", 11), text_color=color).pack(side="left", padx=15)
        ctk.CTkButton(item, text="View", width=80, height=28, fg_color="#1f6aa5",
                      command=lambda n=name: messagebox.showinfo("Photo Album", f"Opening album: {n}")).pack(side="right", padx=10, pady=10)

class CloudVaultPanel(ctk.CTkFrame):
    def __init__(self, parent, back_callback):
        super().__init__(parent, fg_color="transparent")
        self.back_callback = back_callback
        self._build()

    def _build(self):
        top_bar = ctk.CTkFrame(self, fg_color="#1f4788", height=60, corner_radius=10)
        top_bar.pack(fill="x", pady=(0, 10))
        top_bar.pack_propagate(False)

        ctk.CTkButton(top_bar, text="← Back", width=80, fg_color="transparent",
                      hover_color="#2c5aa0", command=self.back_callback).pack(side="left", padx=10, pady=10)
        ctk.CTkLabel(top_bar, text="☁️  PRIVATE CLOUD VAULT",
                     font=("Arial", 18, "bold"), text_color="white").pack(side="left", padx=10)

        content = ctk.CTkFrame(self, fg_color="#0d1b2a", corner_radius=10)
        content.pack(fill="both", expand=True)

        sidebar = ctk.CTkFrame(content, width=240, fg_color="#0a192f", corner_radius=10)
        sidebar.pack(side="left", fill="y", padx=(10, 5), pady=10)
        sidebar.pack_propagate(False)

        ctk.CTkLabel(sidebar, text="Storage Sections", font=("Arial", 13, "bold"), text_color="#3498db").pack(pady=15)
        
        sections = [
            ("🔒 Secure Business", "hidden"),
            ("📋 Work Documents", "hidden"),
            ("🏗️ Infrastructure Files", "hidden"),
            ("📸 Media Library", "visible"),
            ("📄 Personal Docs", "visible"),
            ("🎬 Video Archive", "visible"),
            ("🎵 Audio Files", "visible"),
        ]
        
        for section_name, visibility in sections:
            badge = "[HIDDEN]" if visibility == "hidden" else ""
           
            ctk.CTkButton(sidebar, text=f"{section_name} {badge}", fg_color="transparent", text_color="#aab7b8",
                          anchor="w", hover_color="#1a2b3c", height=38,
                          command=lambda s=section_name: self._show_section(s)).pack(fill="x", padx=10, pady=4)

        scroll = ctk.CTkScrollableFrame(content, fg_color="transparent")
        scroll.pack(side="right", fill="both", expand=True, padx=(5, 10), pady=10)
        self.scroll_frame = scroll
        
        self._show_section("📸 Media Library")

    def _show_section(self, section_name):
        for widget in self.scroll_frame.winfo_children():
            widget.destroy()
        
        section_data = {
            "🔒 Secure Business": [
                ("Financial_Reports_2025.xlsx", "2.4 MB", "#c0392b"),
                ("Client_Contracts_Folder", "15.6 MB", "#c0392b"),
                ("Strategic_Plans_2026.docx", "3.2 MB", "#c0392b"),
                ("Partnership_Agreements", "8.9 MB", "#c0392b"),
            ],
            "📋 Work Documents": [
                ("Project_Status_Reports", "5.7 MB", "#e74c3c"),
                ("Meeting_Minutes_2026.zip", "4.2 MB", "#e74c3c"),
                ("Team_Performance_Data", "2.1 MB", "#e74c3c"),
                ("Training_Materials", "12.3 MB", "#e74c3c"),
            ],
            "🏗️ Infrastructure Files": [
                ("Server_Configs.tar.gz", "1.8 MB", "#3498db"),
                ("Network_Topology.pdf", "4.5 MB", "#3498db"),
                ("Database_Backups", "256.7 MB", "#3498db"),
                ("System_Architecture.dwg", "7.3 MB", "#3498db"),
            ],
            "📸 Media Library": [
                ("Company_Events_2025", "456.2 MB", "#27ae60"),
                ("Team_Photos", "234.5 MB", "#27ae60"),
                ("Product_Showcase", "189.3 MB", "#27ae60"),
                ("Marketing_Assets", "567.8 MB", "#27ae60"),
            ],
            "📄 Personal Docs": [
                ("Resume_Portfolio.pdf", "2.1 MB", "#f39c12"),
                ("Certificates_Folder", "1.5 MB", "#f39c12"),
                ("Tax_Documents_2025", "3.4 MB", "#f39c12"),
                ("Legal_Agreements", "2.8 MB", "#f39c12"),
            ],
            "🎬 Video Archive": [
                ("Training_Videos.mp4", "234.5 MB", "#9b59b6"),
                ("Presentations_Recorded", "456.7 MB", "#9b59b6"),
                ("Event_Coverage_2025", "678.9 MB", "#9b59b6"),
            ],
            "🎵 Audio Files": [
                ("Podcasts_Library", "123.4 MB", "#1abc9c"),
                ("Meeting_Recordings.zip", "234.5 MB", "#1abc9c"),
                ("Voice_Notes", "45.6 MB", "#1abc9c"),
            ],
        }
        
        section_title = ctk.CTkLabel(self.scroll_frame, text=section_name, 
                                     font=("Arial", 18, "bold"), text_color="white")
        section_title.pack(anchor="w", padx=10, pady=(10, 20))
        
        if section_name in section_data:
            for item_name, size, color in section_data[section_name]:
                self._add_storage_item(item_name, size, color)
    
    def _add_storage_item(self, name, size, color):
        item = ctk.CTkFrame(self.scroll_frame, fg_color="#0a192f", height=60, corner_radius=8)
        item.pack(fill="x", pady=4)
        item.pack_propagate(False)
        
        ctk.CTkLabel(item, text=name, font=("Arial", 13, "bold"), text_color="white").pack(side="left", padx=15, pady=10)
        ctk.CTkLabel(item, text=size, font=("Arial", 11), text_color=color).pack(side="left", padx=15)
        ctk.CTkButton(item, text="Access", width=80, height=28, fg_color="#1f6aa5",
                      command=lambda n=name: messagebox.showinfo("Cloud Storage", f"Accessing: {n}")).pack(side="right", padx=10, pady=10)

class FurnitureBizPanel(ctk.CTkFrame):
    def __init__(self, parent, back_callback):
        super().__init__(parent, fg_color="transparent")
        self.back_callback = back_callback
        self._build()

    def _build(self):
        top_bar = ctk.CTkFrame(self, fg_color="#2c3e50", height=60, corner_radius=10)
        top_bar.pack(fill="x", pady=(0, 10))
        top_bar.pack_propagate(False)

        ctk.CTkButton(top_bar, text="← Back", width=80, fg_color="transparent",
                      hover_color="#34495e", command=self.back_callback).pack(side="left", padx=10, pady=10)
        ctk.CTkLabel(top_bar, text="🛋️ MAIN STREET FURNITURE CO.",
                     font=("Arial", 18, "bold"), text_color="white").pack(side="left", padx=10)

        content = ctk.CTkFrame(self, fg_color="#0d1b2a", corner_radius=10)
        content.pack(fill="both", expand=True)

        sidebar = ctk.CTkFrame(content, width=160, fg_color="#0a192f", corner_radius=10)
        sidebar.pack(side="left", fill="y", padx=(10, 5), pady=10)
        sidebar.pack_propagate(False)

        ctk.CTkLabel(sidebar, text="Categories", font=("Arial", 13, "bold"), text_color="#3498db").pack(pady=15)
        for cat in ["Living Room", "Bedroom", "Dining", "Office", "Clearance"]:
            ctk.CTkButton(sidebar, text=cat, fg_color="transparent", text_color="#aab7b8",
                          anchor="w", hover_color="#1a2b3c", height=32).pack(fill="x", padx=8, pady=2)

        scroll = ctk.CTkScrollableFrame(content, fg_color="transparent")
        scroll.pack(side="right", fill="both", expand=True, padx=(5, 10), pady=10)

        inventory_data = [
            ("Velvet Sectional", "$1,299.00", "In Stock (4)"),
            ("Oak Dining Table", "$850.00", "Low Stock (1)"),
            ("Memory Foam Queen", "$599.99", "In Stock (12)"),
            ("Ergo Office Chair", "$245.00", "Out of Stock"),
            ("Mid-Century Lamp", "$89.00", "In Stock (25)")
        ]
        for name, price, stock in inventory_data:
            self._add_item_row(scroll, name, price, stock)

    def _add_item_row(self, parent, name, price, stock):
        row = ctk.CTkFrame(parent, fg_color="#0a192f", height=55, corner_radius=8)
        row.pack(fill="x", pady=4)
        row.pack_propagate(False)

        ctk.CTkLabel(row, text=name, font=("Arial", 13, "bold"), text_color="white").pack(side="left", padx=15)
        color = "#27ae60" if "In Stock" in stock else ("#e67e22" if "Low" in stock else "#e74c3c")
        ctk.CTkLabel(row, text=stock, font=("Arial", 11), text_color=color).pack(side="left", padx=15)
        ctk.CTkButton(row, text=f"Sell {price}", width=100, height=28, fg_color="#1f6aa5",
                      command=lambda n=name: messagebox.showinfo("POS System", f"Opening checkout for: {n}")).pack(side="right", padx=10)

class BlueFinBankPanel(ctk.CTkFrame):
    def __init__(self, parent, current_user, back_callback, logout_callback):
        super().__init__(parent, fg_color="transparent")
        self.back_callback = back_callback
        self.logout_callback = logout_callback
        self.current_user = current_user
        self._build()

    def _build(self):
        top_bar = ctk.CTkFrame(self, fg_color="#0d253f", height=60, corner_radius=10)
        top_bar.pack(fill="x", pady=(0, 10))
        top_bar.pack_propagate(False)

        ctk.CTkButton(top_bar, text="← Back", width=80, fg_color="transparent",
                      hover_color="#1a3a5c", command=self.back_callback).pack(side="left", padx=10, pady=10)
        ctk.CTkLabel(top_bar, text="🏦 BLUE FIN PRIVATE WEALTH",
                     font=("Arial", 18, "bold"), text_color="#3498db").pack(side="left", padx=10)
        ctk.CTkLabel(top_bar, text=f"Node: MO-USA | {datetime.now().strftime('%H:%M')}",
                     text_color="gray", font=("Arial", 11)).pack(side="right", padx=15)

        layout = ctk.CTkFrame(self, fg_color="transparent")
        layout.pack(fill="both", expand=True)

        main_content = ctk.CTkFrame(layout, fg_color="#0a192f", corner_radius=10)
        main_content.pack(side="left", fill="both", expand=True, padx=(0, 5))

        ctk.CTkLabel(main_content, text=f"Welcome, {self.current_user}",
                     font=("Arial", 22, "bold")).pack(anchor="w", padx=20, pady=(20, 5))

        card = ctk.CTkFrame(main_content, fg_color="#1f6aa5", height=100, corner_radius=12)
        card.pack(fill="x", padx=20, pady=10)
        card.pack_propagate(False)
        ctk.CTkLabel(card, text="TOTAL ASSETS", font=("Arial", 11, "bold")).pack(pady=(12, 0))
        ctk.CTkLabel(card, text="$1,245,800.50", font=("Arial", 30, "bold")).pack()

        ctk.CTkLabel(main_content, text="Recent Transactions",
                     font=("Arial", 14, "bold")).pack(anchor="w", padx=20, pady=(15, 5))

        ledger = ctk.CTkTextbox(main_content, fg_color="#0d253f", font=("Courier", 12), corner_radius=8)
        ledger.pack(fill="both", expand=True, padx=20, pady=(0, 10))

        log_data = [
            "2026-03-23 | Wire Transfer IN      | +$15,000.00",
            "2026-03-22 | Server Hosting Fee    |  -$1,240.20",
            "2026-03-21 | SharkShell Premium    |   -$199.99",
            "2026-03-21 | Dividends Received    |  +$2,100.00",
        ]
        ledger.insert("0.0", "\n".join(log_data))
        ledger.configure(state="disabled")

        ctk.CTkButton(main_content, text="SECURE TERMINAL LOGOUT", fg_color="#c0392b",
                      hover_color="#a93226", command=self.logout_callback).pack(pady=(0, 15))

        if self.current_user == "admin":
            oversight_panel = ctk.CTkFrame(layout, width=320, fg_color="#0d253f", corner_radius=10)
            oversight_panel.pack(side="right", fill="y", padx=(5, 0))
            oversight_panel.pack_propagate(False)

            ctk.CTkLabel(oversight_panel, text="BANK OVERLOOK NODE", font=("Arial", 14, "bold"), text_color="#3498db").pack(pady=15)
            
            health_frame = ctk.CTkFrame(oversight_panel, fg_color="#050a14", corner_radius=8)
            health_frame.pack(fill="x", padx=10, pady=5)
            
            ctk.CTkLabel(health_frame, text="NETWORK HEALTH INDICATORS", font=("Arial", 10, "bold"), text_color="gray").pack(pady=(5, 0))
            
            self._add_health_stat(health_frame, "Node Latency", "12ms", "#2ecc71")
            self._add_health_stat(health_frame, "Uptime", "99.998%", "#2ecc71")
            self._add_health_stat(health_frame, "Packet Loss", "0.00%", "#2ecc71")
            self._add_health_stat(health_frame, "Sync Status", "Verified", "#3498db")

            ctk.CTkLabel(oversight_panel, text="LIVE TRANSACTION & DATA TICKER", font=("Arial", 10, "bold"), text_color="gray").pack(pady=(20, 0))
            
            self.ticker_box = ctk.CTkTextbox(oversight_panel, fg_color="#050a14", font=("Courier", 10), text_color="#2ecc71", corner_radius=5)
            self.ticker_box.pack(fill="both", expand=True, padx=10, pady=10)
            
            self._update_ticker()

    def _add_health_stat(self, parent, label, value, color):
        row = ctk.CTkFrame(parent, fg_color="transparent")
        row.pack(fill="x", padx=10, pady=2)
        ctk.CTkLabel(row, text=label, font=("Arial", 11)).pack(side="left")
        ctk.CTkLabel(row, text=value, font=("Arial", 11, "bold"), text_color=color).pack(side="right")

    def _update_ticker(self):
        actions = [
            "TX_AUTH: Success", "NODE_PULSE: OK", "ENCRYPT: AES-256", 
            "DATA_SYNC: 100%", "IP_FILTER: Active", "SQL_AUDIT: Clean",
            "MFA_REQUEST: Sent", "VAULT_LOCK: Secure", "TX_VAL: +$4.2k"
        ]
        timestamp = datetime.now().strftime("%H:%M:%S")
        entry = f"> {random.choice(actions)} [{timestamp}]\n"
        
        self.ticker_box.configure(state="normal")
        self.ticker_box.insert("1.0", entry)
        self.ticker_box.configure(state="disabled")
        
        self.after(2500, self._update_ticker)

class SharkShellApp(ctk.CTk):
    def __init__(self):
        super().__init__()
        self.title("SharkShell Enterprise Security")
        self.geometry("1100x800")
        self.configure(fg_color="#050a14")

        self._user_db = {
            "mark123": pbkdf2_sha256.hash("1234"),
            "admin": pbkdf2_sha256.hash("4321")
        }

        self.flagged_users = set()
        self.activity_log = []
        self.honeypot_log = []
        self.failed_attempts = 0
        self.honeypot_enabled = True
        self.current_user = None

        self.root_frame = ctk.CTkFrame(self, corner_radius=20, fg_color="#0a192f")
        self.root_frame.pack(padx=30, pady=30, fill="both", expand=True)

        self._active_panel = None
        self.render_login()

    def _switch_panel(self, build_fn):
        if self._active_panel and self._active_panel.winfo_exists():
            self._active_panel.destroy()
        panel = build_fn()
        panel.pack(fill="both", expand=True, padx=0, pady=0)
        self._active_panel = panel

    def log_event(self, action):
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        entry = {"action": action, "time": timestamp}
        self.activity_log.insert(0, entry)
        logging.info(f"ACTION LOGGED: {action}")
        if hasattr(self, 'hist_box') and self.hist_box.winfo_exists():
            self.update_history_display()

    def render_login(self):
        def build():
            frame = ctk.CTkFrame(self.root_frame, fg_color="transparent")
            ctk.CTkLabel(frame, text="🦈🔒", font=("Arial", 60)).pack(pady=(60, 5))
            ctk.CTkLabel(frame, text="SHARKSHELL", font=("Arial", 30, "bold"), text_color="#3498db").pack()
            ctk.CTkLabel(frame, text="ENTERPRISE AUTHENTICATION", font=("Arial", 12), text_color="gray").pack(pady=5)
            ctk.CTkLabel(frame, text='"Security That Bites Back"', font=("Arial", 11, "italic"), text_color="gray").pack(pady=3)

            self.user_input = ctk.CTkEntry(frame, placeholder_text="Username", width=300, height=45)
            self.user_input.pack(pady=(40, 10))

            self.pass_input = ctk.CTkEntry(frame, placeholder_text="Password", show="*", width=300, height=45)
            self.pass_input.pack(pady=10)

            self.hp_field = ctk.CTkEntry(frame, width=0, height=0)
            self.hp_field.place(x=-1000, y=-1000)

            ctk.CTkButton(frame, text="AUTHORIZE SESSION", command=self.handle_login,
                          width=300, height=50).pack(pady=40)
            return frame

        self._switch_panel(build)

    def handle_login(self):
        u, p = self.user_input.get(), self.pass_input.get()
        if self.honeypot_enabled and self.hp_field.get():
            if u: self.flagged_users.add(u)
            self.trigger_deception()
            return

        hashed = self._user_db.get(u)
        if hashed and pbkdf2_sha256.verify(p, hashed):
            if self.failed_attempts > 5:
                self.flagged_users.add(u)

            if u in self.flagged_users:
                self.render_verification_gate(u)
                return
            
            if u == "admin":
                otp = str(random.randint(100000, 999999))
                messagebox.showinfo("Admin Verification Required", f"System Admin MFA Token:\n\n{otp}")
                
                user_otp = ctk.CTkInputDialog(text="Enter 6-digit MFA token for Admin Access:", title="Admin Auth").get_input()
                if user_otp != otp:
                    self.log_event("Admin Login MFA Failed")
                    messagebox.showerror("Denied", "Invalid Administrator Token. Access Denied.")
                    return
                self.log_event("Admin Login MFA Authorized")
                
            self.current_user = u
            self.failed_attempts = 0
            self.log_event(f"{u} Authentication Success")
            self.show_hub()
        else:
            self.failed_attempts += 1
            self.log_event(f"Failed Login Attempt ({self.failed_attempts})")
            
            if self.honeypot_enabled and self.failed_attempts >= 3:
                if u: self.flagged_users.add(u)
                self.trigger_deception()
            else:
                messagebox.showerror("Access Denied", "Credentials rejected by system.")

    def render_verification_gate(self, username):
        def build():
            frame = ctk.CTkFrame(self.root_frame, fg_color="transparent")
            ctk.CTkLabel(frame, text="⚠️ SYSTEM INTEGRITY ALERT", font=("Arial", 24, "bold"), text_color="#e74c3c").pack(pady=20)
            ctk.CTkLabel(frame, text=f"Access restricted for: {username}\nContact an administrator.", wraplength=400).pack(pady=10)
            
            ctk.CTkLabel(frame, text="PRIMARY NODE ID").pack(pady=(20, 5))
            user_answer = ctk.CTkEntry(frame, width=300)
            user_answer.pack(pady=5)

            ctk.CTkLabel(frame, text="ADMIN OVERRIDE", font=("Arial", 12, "bold")).pack(pady=(30, 5))
            admin_pass = ctk.CTkEntry(frame, show="*", width=300)
            admin_pass.pack(pady=5)

            def verify_bypass():
                if pbkdf2_sha256.verify(admin_pass.get(), self._user_db.get("admin")) and user_answer.get().strip():
                    self.flagged_users.discard(username)
                    self.log_event(f"Admin Restored Access for {username}")
                    messagebox.showinfo("Success", "Security flag cleared.")
                    self.failed_attempts = 0
                    self.render_login()
                else:
                    messagebox.showerror("Critical Fail", "Administrative override rejected.")

            ctk.CTkButton(frame, text="UNLOCK ACCOUNT", command=verify_bypass, fg_color="#27ae60", width=300).pack(pady=20)
            ctk.CTkButton(frame, text="Back to Login", command=self.render_login, fg_color="transparent").pack()
            return frame
        self._switch_panel(build)

    def secure_logout(self):
        self.log_event(f"Session terminated for {self.current_user}")
        self.current_user = None
        self.failed_attempts = 0
        self.render_login()

    def show_hub(self):
        def build():
            frame = ctk.CTkFrame(self.root_frame, fg_color="transparent")
            top = ctk.CTkFrame(frame, fg_color="#0d253f", height=55, corner_radius=10)
            top.pack(fill="x", pady=(0, 10))
            top.pack_propagate(False)
            ctk.CTkLabel(top, text="🦈 SHARKSHELL", font=("Arial", 16, "bold"), text_color="#3498db").pack(side="left", padx=20, pady=15)
            ctk.CTkButton(top, text="SIGN OUT", width=90, height=30, fg_color="#c0392b", command=self.secure_logout).pack(side="right", padx=15, pady=10)

            self.tabs = ctk.CTkTabview(frame, fg_color="#0a192f", corner_radius=10)
            self.tabs.pack(fill="both", expand=True)

            if self.current_user == "admin":
                self.tabs.add("Security Logs")
                self.tabs.add("Honeypot Events")
                self.tabs.add("Bank Oversight")
                self.tabs.add("Settings")
                self.hist_box = ctk.CTkTextbox(self.tabs.tab("Security Logs"), font=("Courier", 12), fg_color="#050a14")
                self.hist_box.pack(fill="both", expand=True, padx=10, pady=10)
                self.update_history_display()
                self.hp_box = ctk.CTkTextbox(self.tabs.tab("Honeypot Events"), font=("Courier", 12), fg_color="#050a14")
                self.hp_box.pack(fill="both", expand=True, padx=10, pady=10)
                self.update_honeypot_display()
                ctk.CTkButton(self.tabs.tab("Bank Oversight"), text="Access Blue Fin Node", command=self.open_bank).pack(pady=10)
            else:
                self.tabs.add("Services")
                self.tabs.add("History")
                self.tabs.add("Settings")

                nodes = [
                    ("💰  Blue Fin Private Wealth", lambda: self.mfa_protected_launch("Blue Fin", self.open_bank)),
                    ("☁️  Private Cloud Vault", lambda: self.mfa_protected_launch("Cloud Vault", self.open_cloud)),
                    ("🛋️  Furniture Biz Manager", lambda: self.mfa_protected_launch("Furniture Biz", self.open_furniture)),
                    ("🖼️  Photo Vault", lambda: self.mfa_protected_launch("Photo Vault", self.open_photo_vault)),
                ]
                for name, cmd in nodes:
                    ctk.CTkButton(self.tabs.tab("Services"), text=name, width=350, height=45, anchor="w", command=cmd).pack(pady=8)

                self.hist_box = ctk.CTkTextbox(self.tabs.tab("History"), font=("Courier", 12), fg_color="#050a14")
                self.hist_box.pack(fill="both", expand=True, padx=10, pady=10)
                self.update_history_display()

                def request_portal_action():
                    self.log_event(f"Portal Request submitted by {self.current_user}")
                    messagebox.showinfo("Portal Request", "Your request for another portal has been sent to the administrator.")

                ctk.CTkButton(self.tabs.tab("Settings"), text="Request Another Portal", width=250, command=request_portal_action).pack(pady=(20, 10))
                ctk.CTkButton(self.tabs.tab("Settings"), text="Change Password", width=250, 
                              fg_color="#1f6aa5", command=self.change_mark_password).pack(pady=10)

            def toggle_hp():
                self.honeypot_enabled = bool(hp_switch.get())
                self.log_event(f"Honeypot {'enabled' if self.honeypot_enabled else 'disabled'} by {self.current_user}")
                
            hp_switch = ctk.CTkSwitch(self.tabs.tab("Settings"), text="Enable Honeypot Feature", command=toggle_hp)
            if self.honeypot_enabled:
                hp_switch.select()
            hp_switch.pack(pady=10)

            return frame
        self._switch_panel(build)

    def change_mark_password(self):
        new_pass = ctk.CTkInputDialog(text="Enter new 4-digit password:", title="Change Password").get_input()
        if new_pass and len(new_pass) == 4 and new_pass.isdigit():
            self._user_db[self.current_user] = pbkdf2_sha256.hash(new_pass)
            self.log_event(f"{self.current_user} changed password")
            self.log_event(f"SECURITY ALERT: Credential change for {self.current_user}")
            messagebox.showinfo("Success", "Password updated successfully.")
        else:
            messagebox.showerror("Error", "Invalid password. Must be exactly 4 digits.")

    def mfa_protected_launch(self, service_name, success_callback):
        self.log_event(f"{service_name} MFA Requested")
        otp = str(random.randint(100000, 999999))
        messagebox.showinfo("Secure MFA", f"Verification Token for {service_name}:\n\n{otp}")
        user_otp = ctk.CTkInputDialog(text=f"Enter 6-digit MFA token for {service_name}:", title="Auth").get_input()

        if user_otp == otp:
            self.log_event(f"{service_name} MFA Authorized")
            success_callback()
        else:
            self.log_event(f"{service_name} MFA Failed")
            messagebox.showerror("Denied", "Invalid Token. Access Denied.")

    def open_photo_vault(self):
        self.log_event("Accessed Photo Vault")
        self._switch_panel(lambda: PhotoVaultPanel(self.root_frame, back_callback=self.show_hub))

    def open_cloud(self):
        self.log_event("Accessed Cloud Vault")
        self._switch_panel(lambda: CloudVaultPanel(self.root_frame, back_callback=self.show_hub))

    def open_furniture(self):
        self.log_event("Accessed Furniture Biz Manager")
        self._switch_panel(lambda: FurnitureBizPanel(self.root_frame, back_callback=self.show_hub))

    def open_bank(self):
        self._switch_panel(lambda: BlueFinBankPanel(self.root_frame, self.current_user, self.show_hub, self.secure_logout))

    def update_history_display(self):
        self.hist_box.configure(state="normal")
        self.hist_box.delete("0.0", "end")
        for entry in self.activity_log:
            if "admin" in entry['action'].lower() and self.current_user != "admin":
                continue
            self.hist_box.insert("end", f"{entry['action']:<35} | {entry['time']:<20}\n")
        self.hist_box.configure(state="disabled")

    def update_honeypot_display(self):
        if not hasattr(self, 'hp_box') or not self.hp_box.winfo_exists(): return
        self.hp_box.configure(state="normal")
        self.hp_box.delete("0.0", "end")
        for entry in self.honeypot_log:
            self.hp_box.insert("end", f"[{entry['time']}] {entry['event']}\n")
        self.hp_box.configure(state="disabled")

    def trigger_deception(self):
        def build():
            self.hp_frame = ctk.CTkFrame(self.root_frame, fg_color="#000000")
            self._header = ctk.CTkLabel(self.hp_frame, text="🔓 ROOT ACCESS GRANTED - ENTERPRISE DATABASES EXPOSED", font=("Courier", 14, "bold"), text_color="#f1c40f")
            self._header.pack(pady=15)

            self._console = ctk.CTkTextbox(self.hp_frame, fg_color="#000000", text_color="#2ecc71", font=("Courier", 13))
            self._console.pack(fill="both", expand=True, padx=10, pady=10)
            
          
            self._hp_timeout_id = self.after(15000, self._auto_hp_timeout)
            
            def extract_data():
                if hasattr(self, '_hp_timeout_id'):
                    self.after_cancel(self._hp_timeout_id)
                self.honeypot_log.insert(0, {"event": "PROTOCOL DELTA 9 TRIGGERED", "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")})
                self.log_event(f"CRITICAL: {self.current_user if self.current_user else 'Unknown Intruder'} triggered Protocol Delta 9")
                
                if self._console.winfo_exists():
                    self._console.insert("end", "\n> [!] UNAUTHORIZED EXTRACTION DETECTED.\n")
                    self._console.insert("end", "> [!] FLAGGING ADMINISTRATORS AND TRACKING IP ADDRESS.\n")
                    self._console.insert("end", "> [!] STOP ACTION IMMEDIATELY.\n")
                    self._console.insert("end", "> [!] INITIATING PROTOCOL DELTA 9...\n")
         
                    self._console.see("end")

              
                self.hp_frame.configure(fg_color="#ff0800")
                self._console.configure(fg_color="#ff0800", text_color="#ffffff")
                self._header.configure(text_color="#ffffff", text="🚨 PROTOCOL DELTA 9 ACTIVE - LOCKDOWN INITIATED 🚨")
                self._extract_btn.configure(state="disabled", fg_color="#000000", text_color="#ff0800")
                
               
                self.after(9000, self.activate_delta_protocol)

            self._extract_btn = ctk.CTkButton(self.hp_frame, text="Extract Passwords & Financials (2.4GB)", 
                                              fg_color="#c0392b", hover_color="#a93226", 
                                              font=("Arial", 14, "bold"), height=45, command=extract_data)
            self._extract_btn.pack(pady=20)

            threading.Thread(target=self.run_fake_terminal, daemon=True).start()
            return self.hp_frame
        self._switch_panel(build)
        
    def _auto_hp_timeout(self):
        self.honeypot_log.insert(0, {"event": "Intruder timed out (15s) on honeypot screen", "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")})
        self.render_login()

    def run_fake_terminal(self):
        lines = [
            "Initializing root environment...",
            "Bypassing internal firewall policies...",
            "Loading node clusters...",
            "Fetching unencrypted tables...",
            "Exposed table: tbl_Executive_Passwords found.",
            "Exposed table: tbl_Client_Financial_Records found.",
            "Exposed file: Offshore_Bank_Accounts.csv found.",
            "System ready. Awaiting extraction command..."
        ]
        for line in lines:
            if not getattr(self, '_console', None) or not self._console.winfo_exists(): return
            time.sleep(random.uniform(0.3, 0.8))
            try:
                self._console.insert("end", f"> {line}\n")
                self._console.see("end")
            except:
                break

    def activate_delta_protocol(self):
        self.failed_attempts = 0
        if self.current_user:
            self.flagged_users.add(self.current_user)
            self.current_user = None
        self.render_login()

if __name__ == "__main__":
    app = SharkShellApp()
    app.mainloop()
```
